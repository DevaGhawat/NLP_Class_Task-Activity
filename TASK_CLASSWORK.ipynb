```python
nlp = spacy.load("en_core_web_sm")
```

```python
# 1. Dataset
documents = [
    {"title": "Space Robots", "description": "A space adventure where astronauts travel with intelligent robots."},
    {"title": "Healthy Food", "description": "A guide about healthy food, fruits, vegetables, and balanced diet."},
    {"title": "Football Star", "description": "A young football player trains hard to become a sports champion."},
    {"title": "Robot Future", "description": "Robots and artificial intelligence change the future of human life."},
    {"title": "Ocean Mystery", "description": "Scientists explore the deep ocean and discover a hidden city."},
    {"title": "Alien Mission", "description": "A group of astronauts meet aliens during a dangerous space mission."},
    {"title": "Cooking Life", "description": "A chef prepares healthy meals using fresh ingredients and vegetables."},
    {"title": "Crime City", "description": "A detective solves a mysterious crime in a large city."},
    {"title": "AI Healthcare", "description": "Artificial intelligence helps doctors improve healthcare and patient treatment."},
    {"title": "Adventure Planet", "description": "Explorers travel to a new planet for an exciting space adventure."}
]

df = pd.DataFrame(documents)
```

```python
# 2. Preprocessing
def preprocess(text):
    doc = nlp(text.lower())
    words = []

    for token in doc:
        if not token.is_stop and not token.is_punct and not token.is_space:
            words.append(token.lemma_)

    return " ".join(words)

df["clean_text"] = df["description"].apply(preprocess)
```

```python
# 3. TF-IDF Vectorization
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df["clean_text"])

```

```python
# 4. User Query
query = input("Enter your search query: ")

clean_query = preprocess(query)
query_vector = vectorizer.transform([clean_query])
```

```python

# 5. Cosine Similarity
scores = cosine_similarity(query_vector, tfidf_matrix).flatten()

df["similarity_score"] = scores
```

```python

# 6. Top 3 Results
top_3 = df.sort_values(by="similarity_score", ascending=False).head(3)

print("User Query:", query)
print("Clean Query:", clean_query)
print("\nTop 3 Recommended Results:\n")

for i, row in top_3.iterrows():
    print("Title:", row["title"])
    print("Description:", row["description"])
    print("Similarity Score:", round(row["similarity_score"], 3))
    print("-" * 50)
```
